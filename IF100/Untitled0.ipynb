```python

a = input("Enter the house type (1+1, 2+1, or 3+1): ")
b = float(input("Enter the down payment percentage (10-50): "))
c = int(input("Enter the loan term (5, 10, or 20 years): "))
if a == "1+1":
  downp = 3800000*(b/100)
  loan = 3800000 - downp
  tinterestpaid = loan * 0.15* c
  tpaymentamount = 3800000 + tinterestpaid
  mortgage = (tpaymentamount - downp)/(c*12)
  print("\nHouse Price:             ", format(3800000,'.2f'), "TL.")
  print("Down Payment Amount:     ",format(downp,'.2f'), "TL.")
  print("Remaining Loan Amount:   ",format(loan,'.2f'), "TL.")
  print("Total Interest Paid:     ", format(tinterestpaid,'.2f'), "TL.")
  print("Total Payment Amount:    ",format( tpaymentamount ,'.2f'), "TL.")
  print("Monthly Mortgage Payment:", format(mortgage ,'.2f'), "TL." )
elif a == "2+1":
  downp = 4200000*(b/100)
  loan = 4200000 - downp
  tinterestpaid = loan * 0.15* c
  tpaymentamount = 4200000 + tinterestpaid
  mortgage = (tpaymentamount - downp)/(c*12)
  print("\nHouse Price:             ", format(4200000,'.2f'), "TL.")
  print("Down Payment Amount:     ",format(downp,'.2f'), "TL.")
  print("Remaining Loan Amount:   ",format(loan,'.2f'), "TL.")
  print("Total Interest Paid:     ", format(tinterestpaid,'.2f'), "TL.")
  print("Total Payment Amount:    ",format( tpaymentamount ,'.2f'), "TL.")
  print("Monthly Mortgage Payment:", format(mortgage ,'.2f'), "TL." )
elif a == "3+1":
  downp = 4800000*(b/100)
  loan = 4800000 - downp
  tinterestpaid = loan * 0.15* c
  tpaymentamount = 4800000 + tinterestpaid
  mortgage = (tpaymentamount - downp)/(c*12)
  print("\nHouse Price:             ", format(4800000,'.2f'), "TL.")
  print("Down Payment Amount:     ",format(downp,'.2f'), "TL.")
  print("Remaining Loan Amount:   ",format(loan,'.2f'), "TL.")
  print("Total Interest Paid:     ", format(tinterestpaid,'.2f'), "TL.")
  print("Total Payment Amount:    ",format( tpaymentamount ,'.2f'), "TL.")
  print("Monthly Mortgage Payment:", format(mortgage ,'.2f'), "TL." )
```
